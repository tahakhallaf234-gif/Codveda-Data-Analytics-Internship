## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```python
df = pd.read_csv("/content/1) iris.csv")
```

## Display Dataset

```python
df.head(3)
```

##Dataset Information

```python
df.info()
```

##Dataset Shape

```python
print("Dataset Shape:", df.shape)
```

##Check Missing Values

```python
if df.isnull().sum().sum() == 0:
    print("No missing values found.")
```

## Check Duplicate Rows

```python
print(f"Number of duplicate rows: {df.duplicated().sum()}")
```

## Remove Duplicate Rows

```python
df.drop_duplicates(inplace=True)
print(f"New shape of DataFrame after removing duplicates: {df.shape}")
```

## Final Shape

```python
print("Final Shape:", df.shape)
```

## Check Data Types

```python
print(df.dtypes)
```

## Conclusion

- Dataset loaded successfully.
- No missing values found.
- Duplicate rows removed.
- Dataset is ready for Exploratory Data Analysis.

## Summary Statistics

```python
df.describe()
```

## Mean

```python
print("Mean:")
print(df.mean(numeric_only=True))
```

## Median

```python
print("\nMedian:")
print(df.median(numeric_only=True))
```

## Mode

```python
print("\nMode:")
print(df.mode().iloc[0])
```

## Histogram

```python
df.hist(figsize=(10,8))
plt.suptitle("Histograms of Iris Features")
plt.show()
```

## Boxplot

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df.iloc[:,:4])
plt.title("Boxplots of Iris Features")
plt.show()
```

## Scatter Plot

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x="petal_length",
    y="petal_width",
    hue="species"
)

plt.title("Petal Length vs Petal Width")
plt.show()
```

## Correlation Matrix

```python
corr = df.corr(numeric_only=True)

corr
```

## Heatmap

```python
plt.figure(figsize=(7,5))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Matrix")
plt.show()
```

## Conclusion

- Summary statistics calculated successfully.
- Data distributions visualized.
- Strong positive correlation found between petal length and petal width.
- Iris dataset is clean and suitable for machine learning.
